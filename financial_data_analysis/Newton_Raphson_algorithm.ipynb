```julia
using Distributions,Plots,LinearAlgebra,SpecialFunctions,Interact
```

```julia
 @manipulate for nu in 1:1:100
    dist1 = TDist(nu)
    dist2 = Normal(0,1)
    plot([x->pdf(dist1,x),x->pdf(dist2,x)],
         title = "Probability density function",
         labels = ["Student t","Normal"])
end
```
